```python
import snscrape.modules.twitter as snt
import pandas as pd
```

Scrape

```python
%%time
atributes_container = []

for i, tweet in enumerate(snt.TwitterSearchScraper('the last of us lang:en').get_items()):
  if i==10000:
    break
  atributes_container.append([tweet.date, tweet.likeCount, tweet.replyCount, tweet.retweetCount, tweet.rawContent])
  

tweets_df = pd.DataFrame(atributes_container, columns=['Date Created', 'Number of Likes', 'Replies', 'Retweets', 'Content'])
```
